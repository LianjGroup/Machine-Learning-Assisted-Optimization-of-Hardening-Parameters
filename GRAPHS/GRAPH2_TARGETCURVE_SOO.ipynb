```python
################################
##                            ##
##                            ##
## TARGET FORCE DISPLACEMENT  ##
##      CURVES FOR SOO        ##
##                            ##
##                            ##
################################

import matplotlib.pyplot as plt
import pandas as pd

# Define file paths for desired curves
desiredCurvePath1 = "TARGETS_FOR_SOO/DP1000_25C_SwiftVoce_CHD6_curve1/FD_Curve.csv"
desiredCurvePath2 = "TARGETS_FOR_SOO/DP1000_25C_SwiftVoce_NDBR6_curve1/FD_Curve.csv"
desiredCurvePath3 = "TARGETS_FOR_SOO/DP1000_25C_SwiftVoce_NDBR20_curve1/FD_Curve.csv"
desiredCurvePath4 = "TARGETS_FOR_SOO/DP1000_400C_SwiftVoce_NDBR50_curve1/FD_Curve.csv"

# Read data from CSV files into dataframes
desiredCurve1 = pd.read_csv(desiredCurvePath1)
desiredCurve2 = pd.read_csv(desiredCurvePath2)
desiredCurve3 = pd.read_csv(desiredCurvePath3)
desiredCurve4 = pd.read_csv(desiredCurvePath4)

# Set the font size for the plot
letterSize=18

# Create a new figure with specified size
plt.figure(figsize=(7,7))

# Plot the desired curves with different colors and labels
plt.plot(desiredCurve1['displacement/mm'], desiredCurve1['force/N'], label='CHD6 @ 25C', color='lime')
plt.plot(desiredCurve2['displacement/mm'], desiredCurve2['force/N'], label='NDBR6 @ 25C', color='magenta')
plt.plot(desiredCurve3['displacement/mm'], desiredCurve3['force/N'], label='NDBR20 @ 25C', color ='darkorange')
plt.plot(desiredCurve4['displacement/mm'], desiredCurve4['force/N'], label='NDBR50 @ 400C', color = 'salmon')

# Set plot title and axis labels with the specified font size
plt.title('TARGET F/D CURVES FOR SOO DP1000', fontsize=letterSize)
plt.xlabel('DISP (MM)',fontsize=letterSize)
plt.ylabel('FORCE (N)', fontsize=letterSize)

# Set font size for tick labels and legend
plt.xticks(fontsize=letterSize)
plt.yticks(fontsize=letterSize)
plt.legend(fontsize=letterSize)

```
