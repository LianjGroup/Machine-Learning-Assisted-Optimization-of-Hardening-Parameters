```python
################################
##                            ##
##                            ##
## TARGET FORCE DISPLACEMENT  ##
##   CURVES FOR MOO AT 400C   ##
##                            ##
##                            ##
################################

import matplotlib.pyplot as plt
from ASSETS.PART2_ASSTFUNCT import*

# Define file paths for desired curves
desiredCurvePath1 = "TARGETS_FOR_MOO/DP1000_400C_SwiftVoce_curve1/NDBR50/FD_Curve.csv"
desiredCurvePath2 = "TARGETS_FOR_MOO/DP1000_400C_SwiftVoce_curve1/NDBR6/FD_Curve.csv"
desiredCurvePath3 = "TARGETS_FOR_MOO/DP1000_400C_SwiftVoce_curve1/CHD6/FD_Curve.csv"

# Specify the launch index, R-squared threshold, and read the desired curves
launchIndex = 18
r_squared_threshold = 0.8
desiredCurve1 = pd.read_csv(desiredCurvePath1)
desiredCurve2 = pd.read_csv(desiredCurvePath2)
desiredCurve3 = pd.read_csv(desiredCurvePath3)

# Calculate yielding indices for each desired curve
breakpoint1 = calculate_yielding_index(desiredCurve1['displacement/mm'], desiredCurve1['force/N'], r2_threshold=r_squared_threshold)
breakpoint2 = calculate_yielding_index(desiredCurve2['displacement/mm'], desiredCurve2['force/N'], r2_threshold=r_squared_threshold)
breakpoint3 = calculate_yielding_index(desiredCurve3['displacement/mm'], desiredCurve3['force/N'], r2_threshold=r_squared_threshold)

# Set the font size for the plot
letterSize=12

# Create a new figure with specified size
plt.figure(figsize=(7,7))

# Plot the first desired curve
plt.plot(desiredCurve1['displacement/mm'], desiredCurve1['force/N'], color="blue",label='NDBR50')
plt.axvline(x=desiredCurve1['displacement/mm'][breakpoint1], color='orange', linestyle='--', label='YIELDING POINT')

# Plot the second desired curve
plt.plot(desiredCurve2['displacement/mm'], desiredCurve2['force/N'], color="red", label='NDBR6')
plt.axvline(x=desiredCurve2['displacement/mm'][breakpoint2], color="cyan", linestyle='--')

# Plot the third desired curve
plt.plot(desiredCurve3['displacement/mm'], desiredCurve3['force/N'], color="peru", label='CHD6')
plt.axvline(x=desiredCurve3['displacement/mm'][breakpoint3], color='fuchsia', linestyle='--')

# Set plot title and axis labels with the specified font size
plt.title('TARGET F/D CURVES FOR MOO 400C, DP1000', fontsize=letterSize)
plt.xlabel('DISP (MM)',fontsize=letterSize)
plt.ylabel('FORCE (N)', fontsize=letterSize)

# Set font size for tick labels
plt.xticks(fontsize=letterSize)
plt.yticks(fontsize=letterSize)

# Add a legend with a slightly smaller font size
plt.legend(fontsize=letterSize+3)

# Display the plot
plt.show()
```
