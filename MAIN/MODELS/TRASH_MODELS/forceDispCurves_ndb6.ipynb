```python
import pandas as pd
import matplotlib.pyplot as plt

# Define the file path and folder name
file_path_ndbr50_25c = 'EXP_FD/NDBR6_FD_Curve.csv'
file_path_ndbr50_400c = 'EXP_FD/400c/NDBR6_400C_FD_Curve.csv'
file_path_ndbr50_25c_predicted = 'PREDICTED_FD/ndb6_25c_FD.csv'
file_path_ndbr50_400c_predicted = 'PREDICTED_FD/ndb6_400c_FD.csv'

# Read the data from the CSV file with the correct delimiter and skipping initial spaces
data_ndbr50_25c = pd.read_csv(file_path_ndbr50_25c)
data_ndbr50_400c = pd.read_csv(file_path_ndbr50_400c)
data_ndbr50_25c_predicted = pd.read_csv(file_path_ndbr50_25c_predicted, delimiter=',', skipinitialspace=True)
data_ndbr50_400c_predicted = pd.read_csv(file_path_ndbr50_400c_predicted, delimiter=',', skipinitialspace=True)

# Extract the displacement and force columns
disp_ndbr50_25c = data_ndbr50_25c['displacement/mm']
force_ndbr50_25c = data_ndbr50_25c['force/kN']

disp_ndbr50_400c = data_ndbr50_400c['displacement/mm']
force_ndbr50_400c = data_ndbr50_400c['force/kN']

disp_ndbr50_25c_predicted = data_ndbr50_25c_predicted['Disp']
force_ndbr50_25c_predicted = data_ndbr50_25c_predicted['Force']

disp_ndbr50_400c_predicted = data_ndbr50_400c_predicted['Disp']
force_ndbr50_400c_predicted = data_ndbr50_400c_predicted['Force']

plt.figure(figsize=(10, 6))
plt.plot(disp_ndbr50_25c, force_ndbr50_25c, marker='o', linestyle='-', color='mediumblue', label='25C Experimental')
plt.plot(disp_ndbr50_400c, force_ndbr50_400c, marker='o', linestyle='-', color='firebrick', label='400C Experimental')
plt.plot(disp_ndbr50_25c_predicted, force_ndbr50_25c_predicted, linestyle='-', linewidth=8, color='skyblue', label='25C Predicted')
plt.plot(disp_ndbr50_400c_predicted, force_ndbr50_400c_predicted, linestyle='-', linewidth=8, color='lightcoral', label='400C Predicted')

plt.title('Experimental vs Predicted FD Curve for NDB6')
plt.xlabel('Displacement (mm)')
plt.ylabel('Force (kN)')
plt.legend()
plt.grid(True)

plt.xlim(0, 1.2)
plt.show()
```
