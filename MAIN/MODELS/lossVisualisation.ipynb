```python
import matplotlib.pyplot as plt
import numpy as np

# Load loss values from both files
loss_25C = np.load('loss_values_25C.npy', allow_pickle=True).item()
loss_400C = np.load('loss_values_400C.npy', allow_pickle=True).item()

# Plot the training and validation losses for both temperatures on the same plot
plt.figure(figsize=(8, 6))

# Training loss for 25C
plt.plot(loss_25C['train'], label='Train Loss (25C)', linestyle='-', color='blue', linewidth=3)

# Validation loss for 25C
plt.plot(loss_25C['valid'], label='Validation Loss (25C)', linestyle='-', color='cyan', linewidth=3)

# Training loss for 400C
plt.plot(loss_400C['train'], label='Train Loss (400C)', linestyle='-', color='purple', linewidth=3)

# Validation loss for 400C
plt.plot(loss_400C['valid'], label='Validation Loss (400C)', linestyle='-', color='pink', linewidth=3)

plt.xlabel('Epoch', fontsize=14)
plt.ylabel('Loss', fontsize=14)
plt.title('Training and Validation Loss Over Epochs at 25C and 400C', fontsize=16)
plt.legend(fontsize=12)
plt.grid(True, linestyle='--', alpha=0.7)
plt.show()
```

```python
import matplotlib.pyplot as plt

# R-squared values
r2_values = {
    'NN (25C)': 0.5450,
    'RF (25C)': 0.67,
    'XGB (25C)' : 0.55,
    'NN (400C)': 0.4572,
    'RF (400C)': 0.57,
    'XGB (400C)': 0.48
}

# Create a bar chart
plt.figure(figsize=(8, 6))
plt.bar(r2_values.keys(), r2_values.values(), color=['blue', 'red', 'green', 'blue', 'red', 'green'])
plt.xlabel('Models and Temperatures', fontsize=14)
plt.ylabel('R-squared Value', fontsize=14)
plt.title('R-squared Values of Different Models at Different Temperatures', fontsize=16)
plt.ylim(0, 1)  # Set y-axis limit to [0, 1] for better visualization
plt.grid(axis='y', linestyle='--', alpha=0.7)

# Display the values on top of the bars
for key, value in r2_values.items():
    plt.text(key, value + 0.01, f'{value:.4f}', ha='center', fontsize=10)

plt.show()
```
