```python
import pandas as pd

def loadSimFD(rptFile, save_csv=False):
    simdata = pd.read_fwf(rptFile, header=1, names=['X','Disp','Force'], delim_whitespace=True, delimiter='|', skip_blank_lines=True)
    simdata.dropna(inplace=True)
    display(simdata) 
    sim_disp = simdata['Disp']
    sim_force = simdata['Force']

    if save_csv:
        csv_filename = rptFile.replace('.rpt', '_processed.csv')
        simdata.to_csv(csv_filename, index=False)
        print(f"Processed data saved to: {csv_filename}")

    return sim_disp, sim_force

# File paths
files = [
    "RPT_FILES/chd6_xgb_25c_secondTry.rpt",
    "RPT_FILES/ndb6_xgb_25c_secondTry.rpt",
    "RPT_FILES/ndb20_xgb_25c_secondTry.rpt",
]

# Process each file and save the processed data to CSV
for file in files:
    print(f"Processing file: {file}")
    sim_disp, sim_force = loadSimFD(file, save_csv=True)
```
