```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

def swiftVoceT(x, c1, c2, c3, c4, c5, c6, c7):
    return c1*c2*(x + c3) ** c4  +  (1-c1)*c5 + c6 * (1 - np.exp(-c7 * x))

truePlasticStrain = np.arange(0, 3.0005, 0.0005)
y_line_ideal = swiftVoceT(truePlasticStrain, 0.5, 1300, 2.3 * 10**-14, 0.075, 773.18, 1039.37, 73.94)
y_line_randomForest_25 = swiftVoceT(truePlasticStrain,0.5139736735702103, 1195.916454402472, 0.023765172822753193, 0.2637934652246432, 831.7697967468094, 811.5048884934508, 251.41025788472933)
y_line_randomForest_400 = swiftVoceT(truePlasticStrain,0.5164042857141723, 1265.4283509977704, 0.023072948243025932, 0.23821238259451316, 899.5989875452575, 836.5955221045319, 237.96493593148278)
y_line_nn_best = swiftVoceT(truePlasticStrain, 0.6687430739402771, 1276.127197265625, 0.043472327291965485, 0.4672930836677551, 1467.2874755859375, 411.266357421875, 347.3883361816406) 
y_line_nn_new_best_fixed = swiftVoceT(truePlasticStrain, 0.6687430739402771, 1276.127197265625, 0.043472327291965485, 0.17, 1467.2874755859375, 700, 347.3883361816406) 
y_line_nn_new_best_400c_fixed = swiftVoceT(truePlasticStrain, 0.4973168671131134, 1406.2655029296875, 0.04360615462064743, 0.17, 856.2398681640625, 700, 456.01861572265625)

plt.figure(1)
plt.plot(truePlasticStrain, y_line_ideal, color='black', label='Ideal')
plt.plot(truePlasticStrain, y_line_randomForest_25, color = 'red', label='Random Forest 25C')
plt.plot(truePlasticStrain, y_line_randomForest_400, color = 'cyan', label='Random Forest 400C')
plt.plot(truePlasticStrain, y_line_nn_new_best_fixed, color = 'green', label='NN 25C')
plt.plot(truePlasticStrain, y_line_nn_new_best_400c_fixed, color = 'purple', label='NN 400c')

plt.title('Flowcurve for All Geometries')
plt.ylim(bottom=0, top=2500)
plt.xlabel('True Strain, -')
plt.ylabel('True Stress, MPa')
plt.legend()
plt.show()
```
