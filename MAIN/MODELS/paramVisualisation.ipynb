```python
import matplotlib.pyplot as plt
import numpy as np

# Data for each model and temperature
data = {
    'NN_25C': [0.6687, 1276.13, 0.0435, 0.17, 1467.29, 700, 347.39],
    'RF_25C': [0.7424742474247418, 1265.030503050307, 0.048064406440643984, 0.17, 1048.290829082908, 700, 367.249724972497],
    'XGB_25C': [0.7933314442634583, 970.584716796875, 0.04525449126958847, 0.13882778584957123, 1274.2763671875, 700, 297.65753173828125],
}

# Experimental data
experimental_data = [0.5, 1300, 2.3e-14, 0.075, 773.18, 1039.37, 73.94]

# Create subplots for each parameter (c1 to c7)
fig, axes = plt.subplots(2, 4, figsize=(20, 8))

# Flatten the axes for easy iteration
axes = np.ravel(axes)

# Specify colors
colors = ['blue', 'red', 'green', 'blue', 'red', 'green', 'black']

# Iterate over parameters
for i, parameter in enumerate(['c1', 'c2', 'c3', 'c4', 'c5', 'c6', 'c7']):
    # Iterate over specified models and 'Experimental'
    for j, (model_temp, values) in enumerate(data.items()):
        model, temp = model_temp.split('_')
        axes[i].bar(model_temp, values[i], color=colors[j], alpha=0.7)

    # Plot experimental data
    axes[i].bar('Experimental', experimental_data[i], color='black', alpha=0.7)

    #axes[i].set_xlabel('Models and Temperature', fontsize=22)
    axes[i].set_ylabel(f'Value', fontsize=24)
    axes[i].set_title(f'{parameter}', fontsize=26)
    axes[i].tick_params(axis='x', rotation=45, labelsize=16)  # Rotate x-axis labels and increase tick label size
    axes[i].tick_params(axis='y', labelsize=14)  # Increase tick label size for y-axis


fig.delaxes(axes[-1])

# Add a main title
fig.suptitle('Parameters Across Models and Temperatures at 25C', fontsize=26)

plt.tight_layout(rect=[0, 0.03, 1, 0.95])
plt.show()
```
