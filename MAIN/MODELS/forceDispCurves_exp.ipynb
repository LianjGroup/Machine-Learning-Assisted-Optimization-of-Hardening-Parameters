```python
import pandas as pd
import matplotlib.pyplot as plt

# Define the file path and folder name
file_path_chd6_25c = 'EXP_FD/CHD6_FD_Curve.csv'
file_path_ndbr6_25c = 'EXP_FD/NDBR6_FD_Curve.csv'
file_path_ndbr20 = 'EXP_FD/NDBR20_FD_Curve.csv'
file_path_ndbr50_25c = 'EXP_FD/NDBR50_FD_Curve.csv'

# Read the data from the CSV file
data_chd6 = pd.read_csv(file_path_chd6_25c)
data_ndbr6 = pd.read_csv(file_path_ndbr6_25c)
data_ndbr20 = pd.read_csv(file_path_ndbr20)
data_ndbr50 = pd.read_csv(file_path_ndbr50_25c)

# Extract the displacement and force columns
disp_chd6 = data_chd6['displacement/mm']
force_chd6 = data_chd6['force/kN']

disp_ndbr6 = data_ndbr6['displacement/mm']
force_ndbr6 = data_ndbr6['force/kN']

disp_ndbr20 = data_ndbr20['displacement/mm']
force_ndbr20 = data_ndbr20['force/kN']

disp_ndbr50 = data_ndbr50['displacement/mm']
force_ndbr50 = data_ndbr50['force/kN']

# Create a Force-Displacement curve for 25C 
plt.figure(figsize=(10, 6))
plt.plot(disp_chd6, force_chd6, marker='o', linestyle='-')
plt.plot(disp_ndbr6, force_ndbr6, marker='o', linestyle='-')
plt.plot(disp_ndbr20, force_ndbr20, marker='o', linestyle='-')
plt.plot(disp_ndbr50, force_ndbr50, marker='o', linestyle='-')

plt.title('Experimental Force-Displacement Curve @ 25C')
plt.xlabel('Displacement (mm)')
plt.ylabel('Force (N)')
plt.grid(True)

# Display the plot or save it to a file
plt.show()
```
