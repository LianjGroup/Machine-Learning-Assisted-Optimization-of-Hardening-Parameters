```python
import pandas as pd
import matplotlib.pyplot as plt

# Define the file path and folder name
file_path_ndbr50_25c = 'EXP_FD/NDBR50_FD_Curve.csv'
file_path_ndbr50_400c = 'EXP_FD/400c/NDBR50_400C_FD_Curve.csv'
file_path_ndbr50_25c_predicted = 'PREDICTED_FD/RF/ndb50_rf_25c_processed.csv'
file_path_ndbr50_400c_predicted = 'PREDICTED_FD/RF/ndb50_rf_400c_processed.csv'

# Read the data from the CSV file with the correct delimiter and skipping initial spaces
data_ndbr50_25c = pd.read_csv(file_path_ndbr50_25c)
data_ndbr50_400c = pd.read_csv(file_path_ndbr50_400c)
data_ndbr50_25c_predicted = pd.read_csv(file_path_ndbr50_25c_predicted, delimiter=',', skipinitialspace=True)
data_ndbr50_400c_predicted = pd.read_csv(file_path_ndbr50_400c_predicted, delimiter=',', skipinitialspace=True)

# Extract the displacement and force columns
disp_ndbr50_25c = data_ndbr50_25c['displacement/mm']
force_ndbr50_25c = data_ndbr50_25c['force/kN']

disp_ndbr50_400c = data_ndbr50_400c['displacement/mm']
force_ndbr50_400c = data_ndbr50_400c['force/kN']

disp_ndbr50_25c_predicted = data_ndbr50_25c_predicted['Disp']
force_ndbr50_25c_predicted = data_ndbr50_25c_predicted['Force']

disp_ndbr50_400c_predicted = data_ndbr50_400c_predicted['Disp']
force_ndbr50_400c_predicted = data_ndbr50_400c_predicted['Force']

plt.figure(figsize=(10, 6))
plt.plot(disp_ndbr50_25c, force_ndbr50_25c, marker='o', linestyle='-', color='mediumblue', label='25C Experimental')
plt.plot(disp_ndbr50_400c, force_ndbr50_400c, marker='o', linestyle='-', color='firebrick', label='400C Experimental')
plt.plot(disp_ndbr50_25c_predicted, force_ndbr50_25c_predicted, linestyle='-', linewidth=8, color='skyblue', label='25C Predicted')
plt.plot(disp_ndbr50_400c_predicted, force_ndbr50_400c_predicted, linestyle='-', linewidth=8, color='lightcoral', label='400C Predicted')

plt.title('Experimental vs Predicted FD Curve for NDB50')
plt.xlabel('Displacement (mm)')
plt.ylabel('Force (kN)')
plt.legend()
plt.grid(True)

plt.xlim(0, 2)
plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt

def plot_subplot(ax, title, disp_exp_25c, force_exp_25c, disp_pred_25c, force_pred_25c, disp_exp_400c, force_exp_400c, disp_pred_400c, force_pred_400c, xlim):
    ax.plot(disp_exp_25c, force_exp_25c, linestyle='--', linewidth=8, color='firebrick', label='25C Experimental')
    ax.plot(disp_pred_25c, force_pred_25c, linestyle='-', linewidth=8, color='firebrick', label='25C Predicted')
    ax.plot(disp_exp_400c, force_exp_400c, linestyle='--', linewidth=8, color='lightcoral', label='400C Experimental')
    ax.plot(disp_pred_400c, force_pred_400c, linestyle='-', linewidth=8, color='lightcoral', label='400C Predicted')
    ax.set_title(title)
    ax.set_xlabel('Displacement (mm)')
    ax.set_ylabel('Force (kN)')
    ax.legend()
    ax.grid(True)
    ax.set_xlim(0, xlim)

# Subplot 1
fig, axs = plt.subplots(2, 2, figsize=(15, 10))

# NDBR50
plot_subplot(axs[0, 0], 'NDBR50', disp_ndbr50_25c, force_ndbr50_25c, disp_ndbr50_25c_predicted, force_ndbr50_25c_predicted,
             disp_ndbr50_400c, force_ndbr50_400c, disp_ndbr50_400c_predicted, force_ndbr50_400c_predicted, 2)

# NDBR20
plot_subplot(axs[0, 1], 'NDBR20', disp_ndbr20_25c, force_ndbr20_25c, disp_ndbr20_25c_predicted, force_ndbr20_25c_predicted,
             0, 0, disp_ndbr20_400c_predicted, force_ndbr20_400c_predicted, 1.5)

# NDBR6
plot_subplot(axs[1, 0], 'NDBR6', disp_ndbr6_25c, force_ndbr6_25c, disp_ndbr6_25c_predicted, force_ndbr6_25c_predicted,
             disp_ndbr6_400c, force_ndbr6_400c, disp_ndbr6_400c_predicted, force_ndbr6_400c_predicted, 1.2)

# CHD6
plot_subplot(axs[1, 1], 'CHD6', disp_chd6_25c, force_chd6_25c, disp_chd6_25c_predicted, force_chd6_25c_predicted,
             disp_chd6_400c, force_chd6_400c, disp_chd6_400c_predicted, force_chd6_400c_predicted, 1.2)

fig.suptitle('Force-Displacement Curves based on Random Forest Predictions', fontsize=24)

plt.tight_layout()
plt.show()
```
